# Linear Models ‚Äî Explanation


## üî∑ 1. Simple Linear Regression

**What is it?**  
Models the relationship between one input (X) and one output (Y) as a straight line.

**The Formula:**  
$$
Y = \beta_0 + \beta_1 X + \epsilon
$$

- **Y**: what you want to predict (dependent variable)  
- **X**: the input (independent variable)  
- **Œ≤‚ÇÄ**: intercept (where the line crosses the Y-axis)  
- **Œ≤‚ÇÅ**: slope (how much Y changes when X increases by 1)  
- **Œµ**: error (the difference between predicted and actual)

**How does it learn?**  
Uses Ordinary Least Squares (OLS) ‚Äî finds the line that minimizes the sum of squared errors (SSE):  
$$
SSE = \sum (Y_i - \hat{Y}_i)^2
$$

**Real Life Example:**  
Predicting salary based on years of experience:  
- X = years of experience  
- Y = salary  
- Model: Salary = 30,000 + 5,000 √ó (years)

**Assumptions:**  
- Linearity  
- Independence  
- Homoscedasticity  
- Normality  

**Evaluation Metrics:**  
- R¬≤ (R-squared): How much variance in Y is explained by X (0 to 1)  
- MSE (Mean Squared Error): Lower = better  
- RMSE: Square root of MSE

---

## üî∑ 2. Multiple Linear Regression

**What is it?**  
Extends simple regression to use multiple input variables to predict one output.

**The Formula:**  
$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n + \epsilon
$$

**Real Life Example:**  
Predicting house price using:  
- X‚ÇÅ = size (sq ft)  
- X‚ÇÇ = number of bedrooms  
- X‚ÇÉ = distance from city center  
- Model: Price = 50,000 + 200√ó(size) + 10,000√ó(bedrooms) - 5,000√ó(distance)

**Key Concepts:**  
- Multicollinearity (detected using VIF)  
- Overfitting

**Evaluation:**  
- R¬≤, MSE, RMSE  
- Adjusted R¬≤

---

## üî∑ 3. Ridge Regression (L2 Regularization)

**The Problem it Solves:**  
Coefficients can become large and unstable (overfitting) with many or correlated variables.

**What Ridge Does:**  
Adds a penalty to the loss function:  
$$
\text{Loss} = SSE + \lambda \sum \beta_i^2
$$

- Œª (lambda): regularization strength

**Key Property:**  
Ridge shrinks coefficients but never makes them exactly zero.

**When to Use:**  
- Many correlated predictors  
- Reduce overfitting but keep all features

---

## üî∑ 4. Lasso Regression (L1 Regularization)

**What is it?**  
Similar to Ridge, but uses absolute values:  
$$
\text{Loss} = SSE + \lambda \sum |\beta_i|
$$

**Key Property ‚Äî Sparsity:**  
Lasso can push coefficients to exactly zero (feature selection).

**Ridge vs Lasso:**

| Ridge | Lasso |
|-------|-------|
| Penalty: Œ≤¬≤ | Penalty: |Œ≤| |
| Zeros out variables? No | Yes |
| Best for: Many small effects | Few important variables |

**When to Use:**  
When only a few variables matter and you want the model to find them.

---

## üî∑ 5. Elastic Net

**What is it?**  
Combination of Ridge and Lasso:  
$$
\text{Loss} = SSE + \lambda_1 \sum |\beta_i| + \lambda_2 \sum \beta_i^2
$$

- Œ± = 1 ‚Üí pure Lasso  
- Œ± = 0 ‚Üí pure Ridge  
- 0 < Œ± < 1 ‚Üí mix of both

**When to Use:**  
- Many correlated variables  
- Handles correlated groups better than Lasso alone

---

## üî∑ 6. Logistic Regression

**What is it?**  
Classification model (not regression). Predicts probability of belonging to a category.

**The Formula:**  
$$
P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + ...)}}
$$

**Decision Boundary:**  
- If P > 0.5 ‚Üí class 1  
- If P ‚â§ 0.5 ‚Üí class 0

**Real Life Example:**  
Predicting spam email (1 = spam, 0 = not spam)

**Loss Function:**  
Uses Log Loss (Binary Cross-Entropy):  
$$
\text{Loss} = -\sum [y_i \log(\hat{y}_i) + (1-y_i)\log(1-\hat{y}_i)]
$$

**Evaluation Metrics:**  
- Accuracy  
- Precision  
- Recall  
- F1 Score  
- AUC-ROC

**Extensions:**  
- Multinomial Logistic Regression  
- Ordinal Logistic Regression

---

## üî∑ 7. Polynomial Regression

**What is it?**  
Relationship between X and Y is curved, but model is linear in coefficients.

**The Formula:**  
$$
Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \beta_3 X^3 + ... + \epsilon
$$

**Real Life Example:**  
Predicting plant growth over time.

**Warning:**  
Higher degree = more risk of overfitting. Use cross-validation.

---

## üî∑ 8. Bayesian Linear Regression

**What is it?**  
Finds a distribution of possible lines, incorporating uncertainty.

**Key Ideas:**  
- Prior  
- Likelihood  
- Posterior

**Why It's Useful:**  
- Works well with small datasets  
- Gives confidence intervals  
- Incorporate domain knowledge

---

## üî∑ 9. Generalized Linear Models (GLM)

**What is it?**  
Extends linear regression to handle different types of outcome variables using a link function.

**Three Components:**  
- Random component (distribution of Y)  
- Systematic component (linear predictor)  
- Link function

**Common GLMs:**

| Model                | Distribution | Link Function | Use Case            |
|----------------------|--------------|---------------|---------------------|
| Linear Regression    | Normal       | Identity      | Continuous outcomes |
| Logistic Regression  | Binomial     | Logit         | Binary outcomes     |
| Poisson Regression   | Poisson      | Log           | Count data          |
| Gamma Regression     | Gamma        | Log/Inverse   | Positive continuous |

**Example ‚Äî Poisson Regression:**  
Predicting number of calls a call center receives per hour.

---

## üî∑ The Big Picture ‚Äî How They All Connect

## üî∑ Linear Regression Model Family Tree

```
Linear Regression (base)
        ‚îÇ
        ‚îú‚îÄ‚îÄ Add more variables ‚Üí Multiple Linear Regression
        ‚îÇ         ‚îÇ
        ‚îÇ         ‚îú‚îÄ‚îÄ Add L2 penalty ‚Üí Ridge
        ‚îÇ         ‚îú‚îÄ‚îÄ Add L1 penalty ‚Üí Lasso
        ‚îÇ         ‚îî‚îÄ‚îÄ Add both ‚Üí Elastic Net
        ‚îÇ
        ‚îú‚îÄ‚îÄ Curve the features ‚Üí Polynomial Regression
        ‚îú‚îÄ‚îÄ Add probability framework ‚Üí Bayesian Linear Regression
        ‚îú‚îÄ‚îÄ Change outcome type ‚Üí GLM
        ‚îÇ         ‚îî‚îÄ‚îÄ Binary outcome ‚Üí Logistic Regression
```

---

## üî∑ When to Use Which?

| Situation | Recommended Model |
|-----------|-------------------|
| One predictor, continuous outcome | Simple Linear Regression |
| Many predictors, continuous outcome | Multiple Linear Regression |
| Risk of overfitting, keep all features | Ridge |
| Want automatic feature selection | Lasso |
| Many correlated features | Elastic Net |
| Binary classification | Logistic Regression |
| Curved relationship | Polynomial Regression |
| Small data, need uncertainty | Bayesian Linear Regression |
| Count or non-normal outcomes | GLM (Poisson, etc.) |

## üì¶ 1. Imports & Data Loading

### What this section does

- Loads two CSV datasets:
  - `cancer_reg.csv`

- Merges them using a common key: **`geography`**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import (
    LinearRegression, Ridge, Lasso, ElasticNet,
    LogisticRegression, BayesianRidge
)
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import (
    mean_squared_error, r2_score, mean_absolute_error,
    classification_report, confusion_matrix, roc_curve, auc
)
from sklearn.impute import SimpleImputer
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 11
print('‚úÖ Libraries loaded!')
````

## üîß 3. Preprocessing

### What this section does
1. Drops non-numeric / unnecessary columns:
   - `pctsomecol18_24` (too many missing values)
   - `geography` (string)
   - `binnedinc` (categorical bins)
2. Fills remaining missing values with median:
   - `SimpleImputer(strategy='median')`
3. Splits into:
   - `X` features
   - `y` target (`target_deathrate`)
4. Scales features with standardization:
   - `StandardScaler()` makes mean=0 and std=1
5. Train-test split:
   - 80% training, 20% testing

### Extra helper
- Finds the single strongest correlated feature:
  - `BEST_FEAT = corr2.abs().idxmax()`

````python
# Drop high-missing + non-numeric columns
df_clean = df.drop(columns=['pctsomecol18_24', 'geography', 'binnedinc'])

# Impute remaining missing values with median
imputer    = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df_clean), columns=df_clean.columns)

TARGET   = 'target_deathrate'
FEATURES = [c for c in df_imputed.columns if c != TARGET]

X_raw = df_imputed[FEATURES]
y     = df_imputed[TARGET]

# Scale features
scaler   = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_raw), columns=FEATURES)

# Train / Test split
X_tr, X_te, y_tr, y_te = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f'‚úÖ Features : {len(FEATURES)}')
print(f'‚úÖ Train    : {X_tr.shape}')
print(f'‚úÖ Test     : {X_te.shape}')
print(f'‚úÖ Missing  : {df_imputed.isnull().sum().sum()}')

# Helper ‚Äî best single feature
num_df2    = df_imputed.select_dtypes(include=[np.number])
corr2      = num_df2.corr()[TARGET].drop(TARGET)
BEST_FEAT  = corr2.abs().idxmax()
print(f'\nüìå Strongest single predictor: {BEST_FEAT}  (r={corr2[BEST_FEAT]:.3f})')
````

## üî∑ 4. Simple Linear Regression
**One feature ‚Üí one target (straight line)**
$$\hat{y} = \beta_0 + \beta_1 X$$
We use the **single feature most correlated** with the death rate.
### What you compute
- Fit model on train
- Predict on test
- Evaluate:
  - **R¬≤**: how much variance is explained
  - **RMSE**: average error magnitude (penalizes large errors)
  - **MAE**: average absolute error

### Plots
- Scatter + regression line
- Actual vs predicted
- Residual plot (should look random around 0)


````python
# ‚îÄ‚îÄ Simple Linear Regression ‚îÄ‚îÄ
Xs_tr = X_tr[[BEST_FEAT]]
Xs_te = X_te[[BEST_FEAT]]

slr       = LinearRegression()
slr.fit(Xs_tr, y_tr)
y_pred_s  = slr.predict(Xs_te)

r2_s   = r2_score(y_te, y_pred_s)
rmse_s = np.sqrt(mean_squared_error(y_te, y_pred_s))
mae_s  = mean_absolute_error(y_te, y_pred_s)

print('‚îÄ'*45)
print('  Simple Linear Regression Results')
print('‚îÄ'*45)
print(f'  Feature  : {BEST_FEAT}')
print(f'  Intercept: {slr.intercept_:.4f}')
print(f'  Slope    : {slr.coef_[0]:.4f}')
print(f'  R¬≤       : {r2_s:.4f}')
print(f'  RMSE     : {rmse_s:.4f}')
print(f'  MAE      : {mae_s:.4f}')
print('‚îÄ'*45)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Scatter + line
xv = np.linspace(Xs_te[BEST_FEAT].min(), Xs_te[BEST_FEAT].max(), 300).reshape(-1,1)
axes[0].scatter(Xs_te[BEST_FEAT], y_te, alpha=0.35, s=10, color='steelblue', label='Actual')
axes[0].plot(xv, slr.predict(xv), color='red', linewidth=2.5, label='Fit Line')
axes[0].set_xlabel(BEST_FEAT); axes[0].set_ylabel('Death Rate')
axes[0].set_title(f'Simple Linear Regression\nR¬≤={r2_s:.4f}')
axes[0].legend()

# Actual vs Predicted
axes[1].scatter(y_te, y_pred_s, alpha=0.35, s=10, color='purple')
lim = [y_te.min(), y_te.max()]
axes[1].plot(lim, lim, 'r--', linewidth=2, label='Perfect Fit')
axes[1].set_xlabel('Actual'); axes[1].set_ylabel('Predicted')
axes[1].set_title('Actual vs Predicted')
axes[1].legend()

# Residuals
resid = y_te - y_pred_s
axes[2].scatter(y_pred_s, resid, alpha=0.35, s=10, color='teal')
axes[2].axhline(0, color='red', linestyle='--')
axes[2].set_xlabel('Predicted'); axes[2].set_ylabel('Residual')
axes[2].set_title('Residual Plot')

plt.suptitle('Simple Linear Regression', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## üî∑ 5. Multiple Linear Regression
**All features ‚Üí one target**
$$\hat{y} = \beta_0 + \beta_1X_1 + \beta_2X_2 + \cdots + \beta_nX_n$$
### Extra: Multicollinearity Check (VIF)
- You compute **VIF** for each feature.
- If **VIF > 10**, it often indicates strong multicollinearity.
- That‚Äôs why Ridge/Lasso can help.

````python
# ‚îÄ‚îÄ Multiple Linear Regression ‚îÄ‚îÄ
mlr       = LinearRegression()
mlr.fit(X_tr, y_tr)
y_pred_m  = mlr.predict(X_te)

r2_m   = r2_score(y_te, y_pred_m)
rmse_m = np.sqrt(mean_squared_error(y_te, y_pred_m))
mae_m  = mean_absolute_error(y_te, y_pred_m)

print('‚îÄ'*45)
print('  Multiple Linear Regression Results')
print('‚îÄ'*45)
print(f'  R¬≤  : {r2_m:.4f}')
print(f'  RMSE: {rmse_m:.4f}')
print(f'  MAE : {mae_m:.4f}')
print('‚îÄ'*45)

coef_df = pd.DataFrame({'Feature': FEATURES, 'Coef': mlr.coef_})
coef_df = coef_df.reindex(coef_df['Coef'].abs().sort_values(ascending=False).index)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Top 15 coefficients
top15 = coef_df.head(15)
colors = ['#d73027' if v < 0 else '#4575b4' for v in top15['Coef']]
axes[0].barh(top15['Feature'], top15['Coef'], color=colors)
axes[0].axvline(0, color='black', lw=0.8)
axes[0].set_title('Top 15 Coefficients (Standardized)')
axes[0].set_xlabel('Coefficient')

# Actual vs Predicted
axes[1].scatter(y_te, y_pred_m, alpha=0.35, s=10, color='steelblue')
lim = [y_te.min(), y_te.max()]
axes[1].plot(lim, lim, 'r--', lw=2, label='Perfect Fit')
axes[1].set_xlabel('Actual'); axes[1].set_ylabel('Predicted')
axes[1].set_title(f'Actual vs Predicted  R¬≤={r2_m:.4f}')
axes[1].legend()

# Residual histogram
resid_m = y_te - y_pred_m
axes[2].hist(resid_m, bins=50, color='purple', edgecolor='white', alpha=0.8)
axes[2].axvline(0, color='red', linestyle='--')
axes[2].set_xlabel('Residual'); axes[2].set_ylabel('Count')
axes[2].set_title('Residual Distribution')

plt.suptitle('Multiple Linear Regression', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## üî∑ 6. Ridge Regression (L2 Regularization)
**Shrinks coefficients but keeps all features**
$$\text{Loss} = SSE + \lambda\sum\beta_i^2$$
Best for high multicollinearity (as we saw in VIF above).

### What your code does
- Creates a list of candidate `alpha` values
- Uses `GridSearchCV` to choose best alpha by CV R¬≤
- Fits Ridge using best alpha
- Plots:
  - coefficient path vs alpha
  - compare Ridge vs MLR coefficients
  - actual vs predicted


````python
# ‚îÄ‚îÄ Ridge Regression ‚îÄ‚îÄ
alphas = np.logspace(-3, 4, 100)

# Cross-validate to find best alpha
ridge_cv = GridSearchCV(Ridge(), {'alpha': alphas}, cv=5, scoring='r2')
ridge_cv.fit(X_tr, y_tr)
best_alpha_r = ridge_cv.best_params_['alpha']

ridge      = Ridge(alpha=best_alpha_r)
ridge.fit(X_tr, y_tr)
y_pred_r   = ridge.predict(X_te)

r2_r   = r2_score(y_te, y_pred_r)
rmse_r = np.sqrt(mean_squared_error(y_te, y_pred_r))
mae_r  = mean_absolute_error(y_te, y_pred_r)

print('‚îÄ'*45)
print('  Ridge Regression Results')
print('‚îÄ'*45)
print(f'  Best Œ± : {best_alpha_r:.4f}')
print(f'  R¬≤     : {r2_r:.4f}')
print(f'  RMSE   : {rmse_r:.4f}')
print(f'  MAE    : {mae_r:.4f}')
print('‚îÄ'*45)

# Coefficient path
coefs_path = np.array([Ridge(alpha=a).fit(X_tr, y_tr).coef_ for a in alphas])

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Coefficient path
for i in range(coefs_path.shape[1]):
    axes[0].plot(alphas, coefs_path[:, i], alpha=0.5, linewidth=0.9)
axes[0].axvline(best_alpha_r, color='red', linestyle='--', lw=2, label=f'Best Œ±={best_alpha_r:.2f}')
axes[0].axhline(0, color='black', lw=0.5)
axes[0].set_xscale('log'); axes[0].set_xlabel('Œ± (log)')
axes[0].set_ylabel('Coefficient'); axes[0].set_title('Ridge ‚Äî Coefficient Path')
axes[0].legend()

# Ridge vs MLR coefficients
comp = pd.DataFrame({'MLR': mlr.coef_, 'Ridge': ridge.coef_}, index=FEATURES)
top10_idx = comp.abs().sum(axis=1).sort_values(ascending=False).head(10).index
comp.loc[top10_idx].plot(kind='bar', ax=axes[1], color=['steelblue','orange'], edgecolor='white')
axes[1].axhline(0, color='black', lw=0.8)
axes[1].set_title('MLR vs Ridge Coefficients (top 10)')
axes[1].set_ylabel('Coefficient')
axes[1].tick_params(axis='x', rotation=45)

# Actual vs Predicted
axes[2].scatter(y_te, y_pred_r, alpha=0.35, s=10, color='orange')
lim = [y_te.min(), y_te.max()]
axes[2].plot(lim, lim, 'r--', lw=2)
axes[2].set_xlabel('Actual'); axes[2].set_ylabel('Predicted')
axes[2].set_title(f'Actual vs Predicted  R¬≤={r2_r:.4f}')

plt.suptitle('Ridge Regression (L2)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## üî∑ 7. Lasso Regression (L1 Regularization)
**Zeros out unimportant features ‚Äî automatic feature selection**
$$\text{Loss} = SSE + \lambda\sum|\beta_i|$$

### What your code does
- GridSearchCV chooses best alpha
- Lasso sets some coefficients exactly to **0**
- You print:
  - how many features were zeroed
  - how many kept

````python
# ‚îÄ‚îÄ Lasso Regression ‚îÄ‚îÄ
alphas_l = np.logspace(-3, 2, 100)

lasso_cv = GridSearchCV(Lasso(max_iter=10000), {'alpha': alphas_l}, cv=5, scoring='r2')
lasso_cv.fit(X_tr, y_tr)
best_alpha_l = lasso_cv.best_params_['alpha']

lasso      = Lasso(alpha=best_alpha_l, max_iter=10000)
lasso.fit(X_tr, y_tr)
y_pred_l   = lasso.predict(X_te)

r2_l     = r2_score(y_te, y_pred_l)
rmse_l   = np.sqrt(mean_squared_error(y_te, y_pred_l))
mae_l    = mean_absolute_error(y_te, y_pred_l)
n_zero   = np.sum(lasso.coef_ == 0)
n_kept   = len(FEATURES) - n_zero

print('‚îÄ'*45)
print('  Lasso Regression Results')
print('‚îÄ'*45)
print(f'  Best Œ±         : {best_alpha_l:.4f}')
print(f'  R¬≤             : {r2_l:.4f}')
print(f'  RMSE           : {rmse_l:.4f}')
print(f'  MAE            : {mae_l:.4f}')
print(f'  Features zeroed: {n_zero}/{len(FEATURES)}')
print(f'  Features kept  : {n_kept}/{len(FEATURES)}')
print('‚îÄ'*45)

# Coefficient path for Lasso
coefs_lpath = np.array([Lasso(alpha=a, max_iter=10000).fit(X_tr, y_tr).coef_ for a in alphas_l])

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Coefficient path
for i in range(coefs_lpath.shape[1]):
    axes[0].plot(alphas_l, coefs_lpath[:, i], alpha=0.5, linewidth=0.9)
axes[0].axvline(best_alpha_l, color='red', linestyle='--', lw=2, label=f'Best Œ±={best_alpha_l:.3f}')
axes[0].axhline(0, color='black', lw=0.5)
axes[0].set_xscale('log'); axes[0].set_xlabel('Œ± (log)')
axes[0].set_ylabel('Coefficient')
axes[0].set_title('Lasso ‚Äî Coefficient Path\n(Lines hitting 0 = eliminated!)')
axes[0].legend()

# Selected features only
sel_df = pd.DataFrame({'Feature': FEATURES, 'Coef': lasso.coef_})
sel_df = sel_df[sel_df['Coef'] != 0].sort_values('Coef')
colors_l = ['#d73027' if v < 0 else '#4575b4' for v in sel_df['Coef']]
axes[1].barh(sel_df['Feature'], sel_df['Coef'], color=colors_l)
axes[1].axvline(0, color='black', lw=0.8)
axes[1].set_title(f'Lasso Selected Features ({n_kept} kept, {n_zero} zeroed)')
axes[1].set_xlabel('Coefficient')

# Actual vs Predicted
axes[2].scatter(y_te, y_pred_l, alpha=0.35, s=10, color='green')
lim = [y_te.min(), y_te.max()]
axes[2].plot(lim, lim, 'r--', lw=2)
axes[2].set_xlabel('Actual'); axes[2].set_ylabel('Predicted')
axes[2].set_title(f'Actual vs Predicted  R¬≤={r2_l:.4f}')

plt.suptitle('Lasso Regression (L1)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## üî∑ 9. Polynomial Regression
**Captures curved, non-linear relationships**
$$\hat{y} = \beta_0 + \beta_1X + \beta_2X^2 + \beta_3X^3 + \cdots$$
We model `medincome` vs death rate with degrees 1‚Äì4 to show underfitting ‚Üí good fit ‚Üí overfitting.

### What your code does
- Splits only `medincome` vs `target`
- Fits degrees [1,2,3,5]
- Reports Train vs Test R¬≤ to show:
  - underfitting (low degree)
  - good fit
  - overfitting (train high, test drops)

````python
# ‚îÄ‚îÄ Polynomial Regression ‚îÄ‚îÄ
POLY_FEAT = 'medincome'
Xp        = df_imputed[[POLY_FEAT]]
yp        = df_imputed[TARGET]

Xp_tr, Xp_te, yp_tr, yp_te = train_test_split(Xp, yp, test_size=0.2, random_state=42)

degrees = [1, 2, 3, 5]
x_line  = np.linspace(Xp[POLY_FEAT].min(), Xp[POLY_FEAT].max(), 300).reshape(-1,1)
colors_ = ['red', 'green', 'orange', 'purple']

fig, axes = plt.subplots(1, 4, figsize=(22, 5))
poly_results = {}

for ax, deg, col in zip(axes, degrees, colors_):
    pipe = make_pipeline(PolynomialFeatures(deg), LinearRegression())
    pipe.fit(Xp_tr, yp_tr)
    r2_tr = r2_score(yp_tr, pipe.predict(Xp_tr))
    r2_te = r2_score(yp_te, pipe.predict(Xp_te))
    poly_results[f'Deg {deg}'] = {'Train R¬≤': r2_tr, 'Test R¬≤': r2_te}

    ax.scatter(Xp[POLY_FEAT], yp, alpha=0.2, s=5, color='steelblue')
    ax.plot(x_line, pipe.predict(x_line), color=col, lw=2.5)
    ax.set_title(f'Degree {deg}\nTrain R¬≤={r2_tr:.3f} | Test R¬≤={r2_te:.3f}')
    ax.set_xlabel(POLY_FEAT); ax.set_ylabel('Death Rate')

plt.suptitle('Polynomial Regression ‚Äî medincome vs Death Rate', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nüìä Train vs Test R¬≤ by Degree:')
print(pd.DataFrame(poly_results).T.to_string())
````

````python
# ‚îÄ‚îÄ Overfitting curve ‚îÄ‚îÄ
train_r2s, test_r2s = [], []
deg_range = range(1, 8)

for deg in deg_range:
    pipe = make_pipeline(PolynomialFeatures(deg), LinearRegression())
    pipe.fit(Xp_tr, yp_tr)
    train_r2s.append(r2_score(yp_tr, pipe.predict(Xp_tr)))
    test_r2s.append(r2_score(yp_te, pipe.predict(Xp_te)))

plt.figure(figsize=(10, 4))
plt.plot(deg_range, train_r2s, 'o-', color='steelblue', label='Train R¬≤')
plt.plot(deg_range, test_r2s,  'o-', color='red',       label='Test R¬≤')
plt.xlabel('Polynomial Degree')
plt.ylabel('R¬≤ Score')
plt.title('Bias-Variance Tradeoff ‚Äî Polynomial Degree')
plt.legend()
plt.tight_layout()
plt.show()
print('üìå When Train R¬≤ keeps rising but Test R¬≤ drops ‚Üí overfitting!')
````

````python
# ‚îÄ‚îÄ Bayesian Linear Regression ‚îÄ‚îÄ
bayes     = BayesianRidge()
bayes.fit(X_tr, y_tr)
y_pred_b, y_std_b = bayes.predict(X_te, return_std=True)

r2_b   = r2_score(y_te, y_pred_b)
rmse_b = np.sqrt(mean_squared_error(y_te, y_pred_b))
mae_b  = mean_absolute_error(y_te, y_pred_b)

print('‚îÄ'*45)
print('  Bayesian Linear Regression Results')
print('‚îÄ'*45)
print(f'  R¬≤               : {r2_b:.4f}')
print(f'  RMSE             : {rmse_b:.4f}')
print(f'  MAE              : {mae_b:.4f}')
print(f'  Mean Uncertainty : {y_std_b.mean():.4f} (std dev of predictions)')
print('‚îÄ'*45)

# Visualize on single feature for clarity
Xb1  = df_imputed[[BEST_FEAT]].values
yb1  = df_imputed[TARGET].values
Xb1s = StandardScaler().fit_transform(Xb1)

b1d  = BayesianRidge()
b1d.fit(Xb1s, yb1)

x_line_b = np.linspace(Xb1s.min(), Xb1s.max(), 300).reshape(-1,1)
y_mean_b, y_std_b1 = b1d.predict(x_line_b, return_std=True)

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Uncertainty band
axes[0].scatter(Xb1s, yb1, alpha=0.2, s=6, color='steelblue', label='Data')
axes[0].plot(x_line_b, y_mean_b, color='red', lw=2.5, label='Mean Prediction')
axes[0].fill_between(x_line_b.ravel(),
                     y_mean_b - 2*y_std_b1,
                     y_mean_b + 2*y_std_b1,
                     alpha=0.3, color='orange', label='95% Uncertainty')
axes[0].set_xlabel(f'{BEST_FEAT} (scaled)')
axes[0].set_ylabel('Death Rate')
axes[0].set_title('Bayesian Regression ‚Äî Uncertainty Band')
axes[0].legend()

# Actual vs Predicted
axes[1].scatter(y_te, y_pred_b, alpha=0.35, s=10, color='darkorange')
lim = [y_te.min(), y_te.max()]
axes[1].plot(lim, lim, 'r--', lw=2)
axes[1].set_xlabel('Actual'); axes[1].set_ylabel('Predicted')
axes[1].set_title(f'Actual vs Predicted  R¬≤={r2_b:.4f}')

# Uncertainty distribution
axes[2].hist(y_std_b, bins=40, color='orange', edgecolor='white', alpha=0.8)
axes[2].axvline(y_std_b.mean(), color='red', linestyle='--',
                label=f'Mean std={y_std_b.mean():.2f}')
axes[2].set_xlabel('Prediction Std Dev'); axes[2].set_ylabel('Count')
axes[2].set_title('Prediction Uncertainty per County')
axes[2].legend()

plt.suptitle('Bayesian Linear Regression', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

$$\log(\mu) = \beta_0 + \beta_1X_1 + \cdots$$
Rate Ratio = exp(Œ≤): >1 increases deaths, <1 decreases deaths.

````python
# ‚îÄ‚îÄ GLM Poisson Regression ‚îÄ‚îÄ
GLM_FEATS = ['incidencerate','medincome','povertypercent',
             'pctbachdeg25_over','pctunemployed16_over','pctpubliccoverage']

y_count = df_imputed['avgdeathsperyear'].astype(int)
Xg_raw  = df_imputed[GLM_FEATS]
Xg_s    = pd.DataFrame(StandardScaler().fit_transform(Xg_raw), columns=GLM_FEATS)
Xg_sm   = sm.add_constant(Xg_s)

poisson_model = sm.GLM(y_count, Xg_sm, family=sm.families.Poisson()).fit()
print(poisson_model.summary())

y_glm_pred = poisson_model.fittedvalues

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Actual vs Predicted
axes[0].scatter(y_count, y_glm_pred, alpha=0.3, s=8, color='steelblue')
mv = max(y_count.max(), y_glm_pred.max())
axes[0].plot([0,mv],[0,mv],'r--', lw=2, label='Perfect Fit')
axes[0].set_xlabel('Actual Death Count'); axes[0].set_ylabel('Predicted')
axes[0].set_title('Poisson GLM ‚Äî Actual vs Predicted')
axes[0].legend()

# Rate Ratios (exp coef)
rr = pd.DataFrame({'Feature': GLM_FEATS,
                   'Rate Ratio': np.exp(poisson_model.params[1:])}).sort_values('Rate Ratio')
cols_glm = ['#d73027' if v < 1 else '#4575b4' for v in rr['Rate Ratio']]
axes[1].barh(rr['Feature'], rr['Rate Ratio'], color=cols_glm)
axes[1].axvline(1, color='red', linestyle='--', label='RR=1 (no effect)')
axes[1].set_title('Rate Ratios exp(Œ≤)\n>1 increases deaths, <1 decreases')
axes[1].set_xlabel('Rate Ratio')
axes[1].legend()

# Residuals
resid_glm = y_count - y_glm_pred
axes[2].hist(resid_glm, bins=50, color='purple', edgecolor='white', alpha=0.8)
axes[2].axvline(0, color='red', linestyle='--')
axes[2].set_xlabel('Residual'); axes[2].set_ylabel('Count')
axes[2].set_title('Poisson GLM ‚Äî Residual Distribution')

plt.suptitle('GLM ‚Äî Poisson Regression', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

## 13. Final Model Comparison
Head-to-head comparison of all regression models on the same test set.

````python
# ‚îÄ‚îÄ Final Comparison ‚îÄ‚îÄ
models_cmp = {
    'Simple Linear'  : (X_tr[[BEST_FEAT]], X_te[[BEST_FEAT]], LinearRegression()),
    'Multiple Linear': (X_tr, X_te, LinearRegression()),
    'Ridge'          : (X_tr, X_te, Ridge(alpha=best_alpha_r)),
    'Lasso'          : (X_tr, X_te, Lasso(alpha=best_alpha_l, max_iter=10000)),
    'Elastic Net'    : (X_tr, X_te, ElasticNet(**best_p, max_iter=10000)),
    'Polynomial(d=2)': (X_tr[[BEST_FEAT]], X_te[[BEST_FEAT]],
                        make_pipeline(PolynomialFeatures(2), LinearRegression())),
    'Bayesian Ridge' : (X_tr, X_te, BayesianRidge()),
}

rows = []
for name, (Xtr_, Xte_, mdl) in models_cmp.items():
    mdl.fit(Xtr_, y_tr)
    preds = mdl.predict(Xte_)
    rows.append({
        'Model': name,
        'R¬≤'   : round(r2_score(y_te, preds), 4),
        'RMSE' : round(np.sqrt(mean_squared_error(y_te, preds)), 4),
        'MAE'  : round(mean_absolute_error(y_te, preds), 4)
    })

cmp_df = pd.DataFrame(rows).sort_values('R¬≤', ascending=False).reset_index(drop=True)
cmp_df.index += 1

print('='*60)
print('  FINAL MODEL COMPARISON ‚Äî Cancer Death Rate Prediction')
print('='*60)
print(cmp_df.to_string())
print('='*60)
print(f"\n Best Model : {cmp_df.iloc[0]['Model']}  (R¬≤={cmp_df.iloc[0]['R¬≤']})")
print(f" 2nd Best   : {cmp_df.iloc[1]['Model']}  (R¬≤={cmp_df.iloc[1]['R¬≤']})")
````

````python
# ‚îÄ‚îÄ Visual comparison ‚îÄ‚îÄ
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, metric, color in zip(axes, ['R¬≤','RMSE','MAE'], ['steelblue','salmon','teal']):
    sorted_df = cmp_df.sort_values(metric, ascending=(metric != 'R¬≤'))
    bars = ax.barh(sorted_df['Model'], sorted_df[metric], color=color, edgecolor='white')
    for bar, val in zip(bars, sorted_df[metric]):
        ax.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
                f'{val:.4f}', va='center', fontsize=9)
    ax.set_title(f'{metric} Comparison\n(lower is better)' if metric != 'R¬≤'
                 else f'{metric} Comparison\n(higher is better)')
    ax.set_xlabel(metric)

plt.suptitle('Final Model Comparison ‚Äî Cancer Death Rate', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````
